# Example 2: Writing Data
First we import the necessary modules and create reader +  writer objects.

```python
from pythermondt.readers import HDF5Reader
from pythermondt.writers import HDF5Writer, AWSWriter

reader = HDF5Reader(source="./example_data/example-data*.hdf5")
writer = HDF5Writer(destination_folder="./example_data")
uploader = AWSWriter(bucket="ffg-bp", destination_folder="example2_writing_data")
```

Now we can load data using the reader object, manipulate it, and write it back to a file using the writer object. The results can be viewed using: [HDF5 Viewer](https://myhdf5.hdfgroup.org/)

```python
for i, container in enumerate(reader):
    container.add_attributes("/MetaData", author="Max Mustermann", version=12.6, owners=["max", "moritz"], ids={"max": 1, "moritz": 2}) 
    writer.write(container, f"test{i}.hdf5")
```

Additionally, the Data can also be uploaded to an AWS S3 Bucket using the AWSWriter object.

```python
for i, container in enumerate(reader):
    container.add_attributes("/MetaData", author="Max Mustermann", version=12.6, owners=["max", "moritz"], ids={"max": 1, "moritz": 2}) 
    uploader.write(container, f"test{i}.hdf5")
```
